```python
def steprequired(step):
    stepsize = []
    if isinstance(step, str):
        if ":" in step:
            start = float(step.split(':')[0])
            stop = float(step.split(':')[1]) + float(step.split(':')[2])
            step = float(step.split(':')[2])
            z = np.arange(start, stop, step).tolist()
            stepsize.append(z)
            return stepsize[0]
        elif "," in step:
            stepsize = [float(i) for i in (step.split(','))]
            return stepsize

    else:
        stepsize.append(step)
        return stepsize
```

```python
def testfunction(test, strtosplit, para):
    testparameters = []
    if len(strtosplit) == len(para):
        testpara = (dict(zip(para, strtosplit)))
    for testnum in range(len(para)):
        testparameters.append(steprequired(testpara.get(para[testnum])))
    print(test)
    print(para)
    print(testparameters)
```

```python
passparameters = []
for i in range(len(df)):
    v = df.loc[i, :].values.tolist()[2:]
    passparameters.append(v)
```

```python
for i in range(len(passparameters)):
    testfunction(TestList[i], passparameters[i], parameters)
```
